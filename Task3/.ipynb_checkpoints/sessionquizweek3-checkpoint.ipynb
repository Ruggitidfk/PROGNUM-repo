```python
#3.1

masses = [1.9891e+30, 1.8986e+27, 
          5.6846e+26, 1.0243e+26, 8.6810e+25,
          5.9736e+24, 4.8685e+24, 6.4185e+23, 
          3.3022e+23, 7.349e+22, 1.25e22]

masses.sort()
ltmoon = []

for x in range(0, len(masses)):
    if masses[x] <= 7.349e+22:
        ltmoon.append(masses[x])

masses = set(masses)
ltmoon = set(ltmoon)

masses = masses.difference(ltmoon)
masses = list(masses)

print(masses)
        
        
```
