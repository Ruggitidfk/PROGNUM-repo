$ ls *[0-9]*.[co]

The square brackets mean any charcter from the first to the last charcter within the brackets, in alphabetical order.

The diffrence between ls *[0-9]*.[co] and ls *[0-9]*.co, is that the latter asks the shell to list all filenames that end in any number between 0 and 9, followed by any other charcaters expect slashes and dots, ending in ".co", while the former wants the shell to list all filenames have all the same attributes except rather than only ending in ".co" it wants it to end in a dot followed by and character between "c" and "o".
